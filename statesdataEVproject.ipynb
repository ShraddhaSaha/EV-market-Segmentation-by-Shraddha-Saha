```python
sns.countplot(x=statesdataEV["subsidy"])
plt.title('Count plot of subsidy')
plt.xlabel('Subsidy')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x=statesdataEV["road tax"])
plt.title('Count plot of road tax')
plt.xlabel('road tax')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(statesdataEV['subsidy'], bins=5, kde=True, color='blue')
plt.title('Distribution of Subsidy')
plt.xlabel('Subsidy')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(statesdataEV['petrol'], bins=20, kde=True, color='pink')
plt.title('Distribution of Petrol Price')
plt.xlabel('Petrol Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(statesdataEV['diesel'], bins=20, kde=True, color='red')
plt.title('Distribution of Diesel Price')
plt.xlabel('Diesel Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(statesdataEV.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot
sns.pairplot(statesdataEV)
plt.show()
```
